```python
import pandas as pd
import os

# Input and output paths
input_file = '/home/charlie/Desktop/Original_Data/Master_FIles/Master_Dataset.csv'
output_dir = '/home/charlie/Desktop/Original_Data/Detailed_Label_Subset/'  # Adjust to your desired path
Detailed_Label = ["C&C-HeartBeat", "C&C-HeartBeat-Attack", 
                  "C&C-PartOfAHorizontalPortScan"]

chunk_size = 10000
subset_size = 800  # Number of rows per subset

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Dictionary to hold dataframes for each label
label_data = {label: pd.DataFrame() for label in Detailed_Label}

try:
    # Read file in chunks
    df_chunks = pd.read_csv(input_file, chunksize=chunk_size)
    for i, chunk in enumerate(df_chunks):
        print(f"Processing Chunk: {i + 1}")
        
        # Strip unwanted characters from column names
        chunk.columns = chunk.columns.str.strip("'").str.strip()
        
        # Check if 'detailed-label' exists in the chunk
        if 'detailed-label' not in chunk.columns:
            print(f"CRITICAL ERROR: 'detailed-label' column not found in chunk {i + 1}.")
            continue
        
        # Process each label
        for label in Detailed_Label:
            # Filter rows matching the label
            filtered_chunk = chunk[chunk['detailed-label'] == label]
            if not filtered_chunk.empty:
                # Append to the corresponding dataframe in the dictionary
                label_data[label] = pd.concat([label_data[label], filtered_chunk])
                
                # Split into 800-row subsets if enough rows are accumulated
                while len(label_data[label]) >= subset_size:
                    subset = label_data[label].iloc[:subset_size]
                    label_data[label] = label_data[label].iloc[subset_size:]  # Keep remaining rows
                    
                    # Define output file path
                    subset_index = len([f for f in os.listdir(output_dir) if f.startswith(label)])
                    output_file = os.path.join(output_dir, f"{label}_{subset_index + 1}.csv")
                    
                    # Save the subset to a CSV file
                    subset.to_csv(output_file, index=False)
                    
except Exception as e:
    print(f"CRITICAL ERROR: {e}")

# Save any remaining data for each label
for label, df in label_data.items():
    if not df.empty:
        subset_index = len([f for f in os.listdir(output_dir) if f.startswith(label)])
        output_file = os.path.join(output_dir, f"{label}_{subset_index + 1}.csv")
        df.to_csv(output_file, index=False)

print("PROCESSING COMPLETE")
```

```python
import pandas as pd
import os

# Input and output paths
input_file = '/home/charlie/Desktop/Original_Data/Master_FIles/Master_Dataset.csv'
output_dir = '/home/charlie/Desktop/Original_Data/Benign/'  # Adjust to your desired path
Detailed_Label = ["Benign"]  # Filter label
chunk_size = 10000
subset_size = 800  # Number of rows per subset

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Dataframe to accumulate rows for "Benign" label
benign_data = pd.DataFrame()

try:
    # Read file in chunks
    df_chunks = pd.read_csv(input_file, chunksize=chunk_size)
    for i, chunk in enumerate(df_chunks):
        print(f"Processing Chunk: {i + 1}")
        
        # Strip unwanted characters from column names
        chunk.columns = chunk.columns.str.strip("'").str.strip()
        
        # Check if 'label' exists in the chunk
        if 'label' not in chunk.columns:
            print(f"CRITICAL ERROR: 'label' column not found in chunk {i + 1}.")
            continue
        
        # Filter rows matching the "Benign" label
        filtered_chunk = chunk[chunk['label'] == "Benign"]
        if not filtered_chunk.empty:
            # Append to benign_data
            benign_data = pd.concat([benign_data, filtered_chunk])
            
            # Split into 800-row subsets if enough rows are accumulated
            while len(benign_data) >= subset_size:
                subset = benign_data.iloc[:subset_size]
                benign_data = benign_data.iloc[subset_size:]  # Keep remaining rows
                
                # Define output file path
                subset_index = len([f for f in os.listdir(output_dir) if f.startswith("Benign")])
                output_file = os.path.join(output_dir, f"Benign_{subset_index + 1}.csv")
                
                # Save the subset to a CSV file
                subset.to_csv(output_file, index=False)
                    
except Exception as e:
    print(f"CRITICAL ERROR: {e}")

# Save any remaining rows in benign_data
if not benign_data.empty:
    subset_index = len([f for f in os.listdir(output_dir) if f.startswith("Benign")])
    output_file = os.path.join(output_dir, f"Benign_{subset_index + 1}.csv")
    benign_data.to_csv(output_file, index=False)

print("PROCESSING COMPLETE")
```
